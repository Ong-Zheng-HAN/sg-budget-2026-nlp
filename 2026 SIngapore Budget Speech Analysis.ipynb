# Set dependencies and functions

```python
import string
import nltk
from nltk.corpus import stopwords
from nltk import FreqDist
import pandas as pd
from nltk.tokenize import word_tokenize
import numpy as np

from lets_plot import *
LetsPlot.setup_html()
```

```python
stop_words = set(stopwords.words("english"))
stop_words.update(["also", "'s", "us", "must", "many"])

def clean_our_text_pythonian(text):
    tokens = word_tokenize(text)
    return " ".join(
        [word for word in tokens 
         if word.lower() not in stop_words 
         and not word.isdigit() 
         and word not in string.punctuation]
    )
```

# Import data

```python
Text = open("Lawrence Wong Speech.txt").read()

Text
```

# Data Wrangling
## Remove stopwords, punctuations, lower text

```python
cleaned_text =\
(
    clean_our_text_pythonian(Text)
)

cleaned_text
```

## Tokenise your text so the machine can process it

```python
tokens=\
(
    nltk
    .word_tokenize(cleaned_text
                  .lower()
                  )
)

tokens
```

# Calculate and model frequency and 20 most common words

```python
# Calculate frequency distribution
fdist = FreqDist(tokens)

# Find the 20 most common words
top_20_words=\
(
    pd
    .DataFrame((fdist.most_common(20)))
)
top_20_words_renamed =\
(
    top_20_words.rename(columns = {0: "words", 1: "volume"})
)
```

# Data visualisation (top 20 keywords used during speech)

```python
top_20_keywords_SG_budget_2026=\
(
    ggplot(top_20_words_renamed,
           aes(x = "words",
               y = "volume")
          )
    + geom_bar(stat = "identity")
    # + coord_flip()
    + labs(title = "Frequency of words in Prime Minister Lawrence Wong 2026 Budget Speech",
           subtitle = "Top 3 words are 'Singapore', 'Support' and 'AI'",
           caption = "Source: Youtube Transcript")
    + geom_segment(x = 2, xend = 2,
                   y = 65, yend = 60, 
                   arrow = arrow(length = 10, 
                                 type='closed'), 
                   color = "red"
                  )
    + geom_segment(x = 1, xend = 1,
                   y = 75, yend = 70, 
                   arrow = arrow(length = 10, 
                                 type='closed'), 
                   color = "red"
                  )
    + geom_segment(x = 0, xend = 0,
                   y = 95, yend = 90, 
                   arrow = arrow(length = 10, 
                                 type='closed'), 
                   color = "red"
                  )
    + theme_classic()
    + ggsize(1700,800)
)
```

# Model the top 3 most used keywords

```python
from nltk.draw.dispersion import dispersion_plot
```

```python
# Based on the top 3 words usage, let's see dispersion too
finance_words_of_interest = ["singapore", "support", "ai"]
```

```python
rows = []

rows =\
    [
        {"index": idx,
         "word": tok}
        for idx, tok
        in enumerate(tokens)
        if tok in finance_words_of_interest
    ]

df_wong_2026 =\
(
    pd
    .DataFrame(rows)
)
```

```python
df_wong_2026
```

# Using <mark>dispersion plot</mark> to find out the top 3 keywords frequency and timeseries. Time elapsed between his usage of those target words. 

```python
dispersion_plot_for_wong_2026 =\
(
    ggplot(df_wong_2026)
    + geom_point(aes(x = "index",
                     y = "word",
                     color = "word",
                     shape = "word"),
                 size = 3
                )
    + labs(title = "Prime Minister Lawrence Wong 2026 Budget Speech",
           subtitle = "PM was basically rapping about AI at one point. Crazy.",
           x = "Token Index (Look at the Intervals)",
           y = "Budget Speech Words of Interest")
    + ggsize(1700,800)
)
```

```python
dispersion_plot_for_wong_2026
```

# Model the keywords on business and family topics

```python
words_of_interest = ["companies", "workers", "families"]
```

```python
rows = []

rows =\
    [
        {"index": idx,
         "word": tok}
        for idx, tok
        in enumerate(tokens)
        if tok in words_of_interest
    ]

df_workervsfamily_2026 =\
(
    pd
    .DataFrame(rows)
)
```

# Using <mark>dispersion plot</mark> to find out the business topic vs family topic frequency and timeseries. Time elapsed between his usage of those target words. 

```python
dispersion_plot_for_busiess_vs_families_2026 =\
(
    ggplot(df_workervsfamily_2026)
    + geom_point(aes(x = "index",
                     y = "word",
                     color = "word",
                     shape = "word"),
                 size = 3
                )
    + labs(title = "Business vs Family Support",
           subtitle = "From this data it seems there is more focus on supporting the Singapore labour market compared to families",
           x = "Token Index (Look at the Intervals)",
           y = "Budget Speech Words of Interest")
    + ggsize(1700,800)
)
```
